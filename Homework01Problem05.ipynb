```python
## SI values found using the astropy library
def unitFix(constant):
    ## workaround so units are shown in pandas DataFrame w/ consistent
    ## scientific notation for all values
    return '%.4E' % constant.si.value * constant.si.unit

speedOfLight = unitFix(const.c) 
lightMonth = unitFix(const.c) * ('%.4E' % (60*60*24*30) * u.s) ## seconds in a month
lightYear = unitFix(const.c) * ( '%.4E' % (60*60*24*365.25) * u.s) ## seconds in a year
parsec = unitFix(const.pc)
kiloparsec = unitFix(const.kpc)
astronomicalUnit = unitFix(const.au)
gravConst = unitFix(const.G)
solarRad = unitFix(const.R_sun)
solarLum = unitFix(const.L_sun)
solarMass = unitFix(const.M_sun)
jupRad = unitFix(const.R_jup)
jupMass = unitFix(const.M_jup)
earthRad = unitFix(const.R_earth)
earthMass = unitFix(const.M_earth)
bohrRad = unitFix(const.a0)
atomicMass = unitFix(const.u)
neutronMass = unitFix(const.m_n)
protonMass = unitFix(const.m_p)
electronMass = unitFix(const.m_e)
electronCharge = unitFix(const.e)
electronChargeCGS =  ('%.4E' % const.e.gauss.to('cm^(3/2)*g^(1/2)/s').value * 
const.e.gauss.to('cm^(3/2)*g^(1/2)/s').unit) ## workaround for units
electronVolt = unitFix(const.e) * u.J / u.C
boltz = unitFix(const.k_B)
planck = unitFix(const.h)
redPlanck = unitFix(const.hbar)

refTable = pd.DataFrame([
    ['Speed of Light', speedOfLight, speedOfLight.cgs],
    ['Lightmonth', lightMonth, lightMonth.cgs],
    ['Lightyear', lightYear, lightYear.cgs],
    ['Parsec', parsec, parsec.cgs],
    ['Kiloparsec', kiloparsec, kiloparsec.cgs],
    ['Astronomical Unit', astronomicalUnit, astronomicalUnit.cgs],
    ['Universal Gravitational Constant', gravConst, gravConst.cgs],
    ['Solar Radius', solarRad, solarRad.cgs],
    ['Solar Luminosity', solarLum, solarLum.cgs],
    ['Solar Mass', solarMass, solarMass.cgs],
    ['Jupiter Radius', jupRad, jupRad.cgs],
    ['Jupiter Mass', jupMass, jupMass.cgs],
    ['Earth Radius', earthRad, earthRad.cgs],
    ['Earth Mass', earthMass, earthMass.cgs],
    ['Bohr Radius', bohrRad, bohrRad.cgs],
    ['Atomic Mass Unit', atomicMass, atomicMass.cgs],
    ['Neutron Mass', neutronMass, neutronMass.cgs],
    ['Proton Mass', protonMass, protonMass.cgs],
    ['Electron Mass', electronMass, electronMass.cgs],
    ['Electron Charge', electronCharge, electronChargeCGS],
    ['Electron-Volt', electronVolt, electronVolt.cgs],
    ['Boltzmann Constant',boltz, boltz.cgs],
    ['Planck Constant', planck, planck.cgs],
    ['Reduced Planck Constant', redPlanck, redPlanck.cgs]
    
], columns=['Constant','SI','CGS'])

display(refTable.style.hide_index())
```
